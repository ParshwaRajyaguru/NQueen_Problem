```python
global N
N = 7


def validPosition(board, row, column):
	for r in range(row):
		if board[r][column] == 1:
			return False


	r = row
	col = column
	while ((r >= 0) & (col < N)):
		if board[r][col] == 1:
			return False
		r = r - 1
		col = col + 1


	i = row
	j = column
	while ((i >= 0) & (j >= 0)):
		if board[i][j] == 1:
			return False
		j = j - 1
		i = i - 1


	return True



def printResult(board):
	for i in range(N):
		for j in range(N):
			if (board[i][j] == 1):
				print ('(' + str(i) + ', ' + str(j) + ')', end = ' ')
	print ()
	print ()

	return



def arrangeNQueue (board, row, column):
	if ((row == N) & (column == 0)):
		printResult(board)
		return

	
	for c in range(N):
		if (validPosition(board, row, c)):
			board[row][c] = 1

			arrangeNQueue(board, row+1, 0)
	
			board[row][c] = 0



# N = int(input("Enter Queue: "))

board = [[0 for x in range(N)] for y in range(N)]

arrangeNQueue (board, 0, 0)
```
